### str.method

```python
df[df.manufacturer.str.startswith('f')] #startswith doesn't support regex
```

```python
df[df.manufacturer.str.match('[af]')] #start with a or f
```

```python
df.query('manufacturer.str.match("[af]")')
```

```python
df[df.manufacturer.str.match('[^adf]')] #ไม่ขึ้นต้นด้วย a|d|f
```

```python
df[df.manufacturer.str.match('[a-f]')] #ขึ้นต้นด้วย a to f
```

```python
df[df.model.str.match('a\d')] #ขึ้นด้วย a ตามด้วยตัวเลข
```

```python
df[df.model.str.match('\d')] #ขึ้นด้วยตัวเลข
```

```python
df[df.model.str.match('.i')] #ตัวที่ 2 เป็นตัว i
```
