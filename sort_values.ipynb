```python
import pandas as pd
```

```python
df = pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/msleep.csv')
df
```

```python
df.sort_values('sleep_total') #น้อยไปมาก
```

```python
df.sort_values('sleep_total',ascending=False,inplace=True) #เรียงจากมากไปน้อย
df
```

```python
df.sort_values(['vore','sleep_total'],inplace=True)
df
```

```python
df.sort_values(['vore','sleep_total'],ascending=[True,False],inplace=True)
df
```

### Sort by index

```python
df2=df.set_index('name')
df2
```

```python
df2.sort_index(inplace=True)
df2
```

```python
df3=df.set_index(['vore','name'])
df3
```

```python
df3.sort_index()
```
